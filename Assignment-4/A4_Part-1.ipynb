```python
import torch
import matplotlib.pyplot as plt
from utils import MELD, ModelTransformer, ModelGRU, train
```

```python
train_dataset = MELD('Data/train_file.json')
val_dataset = MELD('Data/val_file.json')

model1 = ModelTransformer()
model2 = ModelGRU()
```

```python
with torch.autocast(device_type='cuda', dtype=torch.float16):
    train_losses, val_losses, train_f1s, val_f1s = train(train_dataset, val_dataset, model1, num_epochs=10)
```

```python
plt.plot(train_losses)
plt.plot(val_losses)
plt.legend(['train loss', 'val_loss'])
plt.xticks(range(0, 10))
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.grid(True)
plt.show()
```

```python
plt.plot(train_f1s)
plt.plot(val_f1s)
plt.legend(['train f1s', 'val f1s'])
plt.xticks(range(0, 10))
plt.xlabel("Epochs")
plt.ylabel("F1-Scores")
plt.grid(True)
plt.show()
```
